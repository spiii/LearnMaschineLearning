# Get the data

### Prepare the functions

```python
import pandas as pd
import os

# Constants
DATA_DIR_PATH = "../data/"


# Paths to the data

asylum_seekers = "asylum_seekers.csv"
asylum_seekers_monthly = "asylum_seekers_monthly.csv"
demographics = "demographics.csv"
persons_of_concern = "persons_of_concern.csv"
resettlement = "resettlement.csv"
time_series = "time_series.csv"


# Load the data into panda
def load_csv_data(fileFullName):
    dataDirPath = DATA_DIR_PATH
    csv_path = os.path.join(dataDirPath, fileFullName)
    return pd.read_csv(csv_path)
```

### Load the data

```python


data_asylum_seekers = load_csv_data("asylum_seekers.csv")
data_asylum_seekers_monthly = load_csv_data("asylum_seekers_monthly.csv")
data_demographics = load_csv_data("demographics.csv")
data_persons_of_concern = load_csv_data("persons_of_concern.csv")
data_resettlement = load_csv_data("resettlement.csv")
data_time_series = load_csv_data("time_series.csv")
```

### data_persons_of_concern


**IDP**

Intern Vertriebene (auch: Binnenvertriebene, Binnenflüchtlinge oder aus dem Englischen internally displaced people/IDPs) sind Personen, die gewaltsam aus ihrer angestammten und rechtmäßigen Heimat vertrieben wurden, bei ihrer Flucht – im Unterschied zu Flüchtlingen im rechtlichen Sinn – keine Staatsgrenze überschritten haben und im eigenen Land verblieben sind. Gründe für diese interne Vertreibung sind bewaffnete Konflikte, Gewalt, Menschenrechtsverletzungen und Naturkatastrophen.
